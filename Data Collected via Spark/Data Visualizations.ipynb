```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Question 1: What are the top 5 destinations for Citi bikers based on time of departure?

```python
morningData = pd.read_csv("morningData.csv")
afternoonData = pd.read_csv("afternoonData.csv")
nightData = pd.read_csv("nightData.csv")
latenightData = pd.read_csv("latenightData.csv")
```

```python
plt.figure(figsize = (7,5))
sns.barplot(data = morningData, y = "endStationName", x = "Number")
plt.title("Time of Departure: Morning - 5AM to 12PM")
plt.xlabel("Number of Occurences")
plt.ylabel("Destination")
plt.show()
plt.figure(figsize = (7,5))
sns.barplot(data = afternoonData, y = "endStationName", x = "Number")
plt.title("Time of Departure: Afternoon - 12PM to 6PM")
plt.xlabel("Number of Occurences")
plt.ylabel("Destination")
plt.show()
plt.figure(figsize = (7,5))
sns.barplot(data = nightData, y = "endStationName", x = "Number")
plt.title("Time of Departure: Evening - 6PM to 10PM")
plt.xlabel("Number of Occurences")
plt.ylabel("Destination")
plt.show()
plt.figure(figsize = (7,5))
sns.barplot(data = latenightData, y = "endStationName", x = "Number")
plt.title("Time of Departure: Late Night - 10PM to 5AM")
plt.xlabel("Number of Occurences")
plt.ylabel("Destination")
plt.show()
```

### Question 3: What is the distribution of bike uses depending on the time of day for departure/arrival based on type of day and gender? 

```python
maleWeekday = pd.read_csv("maleWeekday.csv")
maleWeekend = pd.read_csv("maleWeekend.csv")
femaleWeekday = pd.read_csv("femaleWeekday.csv")
femaleWeekend = pd.read_csv("femaleWeekend.csv")
```

```python
maleWeekday.dtypes
```

```python
sns.catplot(x = "Hour", y = "Num", hue = "Type", data = maleWeekday, kind = "bar", )
plt.show()
sns.catplot(x = "Hour", y = "Num", hue = "Type", data = maleWeekend, kind = "bar", )
plt.show()
sns.catplot(x = "Hour", y = "Num", hue = "Type", data = femaleWeekday, kind = "bar", )
plt.show()
sns.catplot(x = "Hour", y = "Num", hue = "Type", data = femaleWeekend, kind = "bar", )
plt.show()
```

### Question 4: What are the most common bikes used and their total hours used? 

```python
mostUsedBikes = pd.read_csv("mostUsedBikes.csv")
mostUsedBikes.bikeID = mostUsedBikes.bikeID.astype(str)
sns.barplot(data = mostUsedBikes, x = "bikeID", y = "Num")
plt.show()
```

```python
mostHoursUsedBikes = pd.read_csv("mostHoursUsedBikes.csv")
mostHoursUsedBikes.bikeID = mostHoursUsedBikes.bikeID.astype(str)
sns.barplot(data = mostHoursUsedBikes, x = "bikeID", y = "HoursUsed")
plt.show()
```
